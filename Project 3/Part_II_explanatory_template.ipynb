## Investigation Overview

This notebook aims to translate the exploratory findings into a concise narrative that answers: How do borrower characteristics and loan features on Prosper influence loan outcomes and interest rates?

Executive Summary of Key Insights:

- Loan Amounts skew low: Most loans are under `$10,000`, with few large outliers.

- Income reduces rates marginally: Higher borrower income correlates weakly with lower interest rates.

- Larger loans get better rates: Moderate inverse relationship between loan size and rate.

- Credit grade segmentations: AA/A grades exhibit low rates and high incomes; HR shows high rates and low incomes.

- Term affects outcomes: 12‑month loans have highest paid‑in‑full rates; 36/60 months see higher defaults/charge‑offs.

## Dataset Overview and Executive Summary

We analyzed 113,937 Prosper loan records with 81 features.  Key fields used:

- LoanOriginalAmount (loan size)

- BorrowerRate (interest rate)

- StatedMonthlyIncome (income)

- ProsperRating (Alpha) (credit grade)

- Term (loan duration)

- LoanStatus (loan outcome)

- Below are three polished visualizations that illustrate these insights.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load dataset
df = pd.read_csv('prosperLoanData.csv') 
sns.set_style('whitegrid')
```

## Visualization 1: Loan Amount Distribution

```python
plt.figure(figsize=(8,4))
sns.histplot(df['LoanOriginalAmount'], bins=30, kde=True)
plt.title('Distribution of Loan Original Amount')
plt.xlabel('Loan Amount ($)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

**Comment:** The distribution is right‑skewed, with most loans below $10,000 and a long tail of larger loans.

## Visualization 2: Income vs. Interest Rate

```python
plt.figure(figsize=(8,6))
plt.scatter(df['StatedMonthlyIncome'], df['BorrowerRate'], alpha=1/8)
plt.xscale('log')
plt.title('Borrower Income vs. Interest Rate')
plt.xlabel('Monthly Income (log scale $)')
plt.ylabel('Interest Rate (%)')
plt.tight_layout()
plt.show()
```

**Comment**: A weak negative trend indicates that higher incomes are associated with slightly lower interest rates, supporting targeted income verification.

## Visualization 3: Credit Grade vs. Borrower Rate

```python
plt.figure(figsize=(10,6))
sns.boxplot(
    x='ProsperRating (Alpha)', y='BorrowerRate', data=df,
    order=['AA','A','B','C','D','E','HR']
)
plt.title('Interest Rate by Credit Grade')
plt.xlabel('Credit Grade')
plt.ylabel('Interest Rate (%)')
plt.tight_layout()
plt.show()
```

**Comment:** AA and A grades have the lowest median rates, while HR has the highest.
